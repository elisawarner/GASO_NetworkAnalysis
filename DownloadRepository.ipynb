# Scan to download files
Description: Input an xml link  
Output: Files saved to a folder

```python
import requests
import shutil
from bs4 import BeautifulSoup
```

```python
#url = 'https://h5.hifly78990.top/app-img/'
```

```python
url = 'https://watermelon-prod-s3.s3.ap-east-1.amazonaws.com/'
```

```python
FOLDERNAME = '../Mitsui55/'
```

```python
def download_file(url):
    local_filename = FOLDERNAME + url.split('/')[-1]
    print(local_filename)
    
    if local_filename[-4:] == 'blob':
        new_filename = local_filename[:-4] + '.png'
    else:
        new_filename = local_filename
    
    with requests.get(url, stream=True) as r:
        r.raw.decode_content = True
        with open(new_filename, 'wb') as f:
            shutil.copyfileobj(r.raw, f)

    return new_filename
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text,'xml')
links = soup.find_all('Key')
```

```python
len(links)
```

```python
for data in links:
    try:
        extension = data.get_text()
        if extension[-3:] == 'wgt' or extension[-3:] == 'tml':
            continue
        
        download_file(url + extension)

    except:
        continue
```
